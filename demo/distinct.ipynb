```python
from colorpalettes.colorset import Colorset
from colorpalettes.optimize import ColorsetOptimizationConfig, optimize_colorset
from colorpalettes.random import random_complimentary_colorset
```

```python
tab_colorset = Colorset.from_cmap("tab10")
tab_colorset.print_distance_stats()
```

```python
tab_colorset.plot_all()
```

```python
seed: int | None = None
init = random_complimentary_colorset(
    start=tab_colorset.colors[0],
    n_colors=8,
    n_color_cycle=3,
    random_seed=seed,
)

optimized = optimize_colorset(
    init,
    config=ColorsetOptimizationConfig(
        saturation_valley=(70, 10),
        lightness_valley=(70, 10),
    ),
)
optimized = optimized.sorted_by_lightness()
optimized.print_distance_stats()
optimized.plot_colors();
```

```python
optimized.plot_all();
```
